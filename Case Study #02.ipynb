# 3. Who scores the most on average for math, reading and writing based on

#  Gender

```python
GM=asg.groupby(["gender"]).mean()
GM.mean(axis=1)
```

```python
#Based on gender ,it can conclude that females have scored more than males on average for math, reading and writing.
```

```python
#Based on test preparation course ,it can concluded that the ones who completed the course have scored more than the ones who have attented no course at all.
```

# 4. What do you think about the scoring variation for math, reading and writing

#  Gender

```python
GM=asg.groupby(["gender"]).mean()
print(GM)
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

X = ['MATHS','READING','WRITING']
Y = [63.376448,72.590734,72.467181]
Z = [68.821577,65.545643,63.446058]
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, Y, 0.4, label = 'Female')
plt.bar(X_axis + 0.2, Z, 0.4, label = 'Male')
plt.xticks(X_axis, X)
plt.xlabel("Groups")
plt.ylabel("Groupwise Average")
plt.title("Variation for Math, Reading and Writing scores based on gender")
plt.legend(loc=(1.02,0))
plt.show()
```

```python
#Average scores are higher for males in Maths ,while average scores of females are higher in reading and writing.
```

#  Test preparation course

```python
TPC=asg.groupby(["test preparation course"]).mean()
print(TPC)
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

X = ['MATHS','READING','WRITING']
Y = [69.96648,74.175978,74.684358]
Z = [63.78972,66.417445,64.457944]
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, Y, 0.4, label = 'Completed ',color='orange')
plt.bar(X_axis + 0.2, Z, 0.4, label = 'None',color='pink')
plt.xticks(X_axis, X)
plt.xlabel("Groups")
plt.ylabel("Groupwise Average")
plt.title("Variation for Math, Reading and Writing scores based on test preparation course")
plt.legend(loc=(1.02,0))
plt.show()
```

```python
#Average scores are higher for the students who completed the test preparation course for all the three groups.
```

# ● Check whether all the zones generate the same amount of sales.

Independent variables are sales in Zones A,B,C & D. As we are having 4 independent variables and to compare the mean of them , consider one-way ANOVA.

Null Hypothesis      - The mean of the sales are equal.
Alternate Hypothesis - The mean is not equal.

```python
#One-way ANOVA can be performed by scipy.stats.f_oneway function in scipy
import scipy.stats as stats
from scipy.stats import f_oneway
```

```python
d_N=data.rename(columns={"Zone - A": "A", "Zone - B": "B","Zone - C":"C","Zone - D":"D"})
f,p=stats.f_oneway(d_N['A'],d_N['B'],d_N['C'],d_N['D'])

print('F-statistic=',round(f,4))
print('p-value=',round(p,4))
if p<0.05:
    print('Reject null Hypothesis.')
else:
    print('Accept null hypothesis.')
```

INSIGHT
The Null Hypothesis is rejected.
Hence we can conclude that all the zones does not generate the same amount of sales.
